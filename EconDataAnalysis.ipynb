```python
unemp_df = fred.search('unemployment state', filter=('frequency', 'Monthly'))
# Creates unemp_df which searches fred for unemployment by state
# Specifically filters to find frequency as monthly (data will be monthly)
unemp_df = unemp_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
# Pulls data from only when there is seasonal adjustment and in percent
unemp_df = unemp_df.loc[unemp_df['title'].str.contains('Unemployment Rate')]
# String filter for only when the title includes 'Unemployment Rate'
```

```python
all_results = []
# List which stores all the results (data) from for loop

for myid in unemp_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_results.append(results)
# Loops through every state ID 
# Then pulls all unemployment data from each state
# The unemployment data is then framed and labeled
# Those results are then stored in all_results
uemp_results = pd.concat(all_results, axis=1)
# Results
```

```python
cols_to_drop = []
# List which stores all the names of columns to get rid of

for i in uemp_results:
    if len(i) > 4:
        cols_to_drop.append(i)
uemp_results = uemp_results.drop(columns=cols_to_drop, axis=1)
# Loops through each column name
# If the length is more than 4, it gets added to cols_to_crop
# Then it is remoced from uemp_results
```

```python
uemp_states = uemp_results.copy()
uemp_states = uemp_states.dropna()
id_to_state = unemp_df['title'].str.replace('Unemployment Rate in ','').to_dict()
#Cleaning code to remove UNRATE and any null values
uemp_states.columns = [id_to_state[c] for c in uemp_states.columns]
# Uses for loop such that states are labeled better
```

```python
px.line(uemp_states, 
              width=1400, 
              height=600,
              title='Unemployment Rate by State', 
              labels={'value': 'Unemployment %',   
                      'index': 'Year'})
# Plots unemployment rates
# Changes width, height, title, and labels
```

```python
uemp_states.loc[uemp_states.index == '2025-09-01'].T.sort_values('2025-09-01').plot(
    kind='barh', figsize=(8,12), title='Unemployment Rate by State: September 1 2025'
)
plt.show()
# Locates uemp_states values at the date 2025-09-01
# Sorts values from high to low and then plots a bar graph
```

```python
part_df = fred.search('participation rate state', filter=('frequency', 'Monthly'))
part_df = part_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
# This code is taken from previous lines
```

```python
part_df['title'].str.replace('Labor Force Participation Rate for ', '')

all_results = []

for myid in part_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_results.append(results)
part_states = pd.concat(all_results, axis=1)
part_id_to_state = part_df['title'].str.replace('Labor Force Participation Rate for ', '').to_dict()
part_states = part_states[[c for c in part_states.columns if c in part_id_to_state]]
part_states.columns = [part_id_to_state[c] for c in part_states.columns]
part_states.columns = part_states.columns.str.strip()
# Taken from previous lines
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax2 = ax.twinx()

uemp_states.loc['2020':'2021', 'California'].plot(
    ax=ax, label='Unemployment', color=color_pal[0]
)
part_states.loc['2020':'2021', 'California'].plot(
    ax=ax2, label='Participation', color=color_pal[1]
)

ax.set_ylabel('Unemployment (%)')
ax2.set_ylabel('Participation (%)')
ax2.grid(False)

fig.legend(loc='upper right')
plt.show()
```
